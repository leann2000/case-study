# ● The mean sales generated by each zone.

```python
print('The mean sale generated by Zone-A:',data['Zone - A'].mean())
print('The mean sale generated by Zone-B:',data['Zone - B'].mean())
print('The mean sale generated by Zone-C:',data['Zone - C'].mean())
print('The mean sale generated by Zone-D:',data['Zone - D'].mean())
```

# ● Total sales generated by all the zones for each month.

```python
data['total_sales'] = data['Zone - A']+data['Zone - B']+data['Zone - C']+data['Zone - D']
```

```python
data.loc[:,['Month','total_sales']]
```

# ● Check whether all the zones generate the same amount of sales.

H0 : All the zones have the same amount of sales. 

H1 : Atleast one zone has a difference in the amount of sales.

The level of significance is 0.05

```python
f_stat,p_value=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
print('The value of F statistics is',f_stat)
print('The p value is',p_value)
```

```python
if p_value < 0.05:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')
```

Hence there is a significant difference in the amount of sales in each zones.
